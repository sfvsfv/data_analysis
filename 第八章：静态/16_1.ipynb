```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([50, 56, 70, 75])

plt.pie(y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([50, 56, 70, 75])

labels = ["橘子", "苹果", "香蕉", "哈密瓜"]
plt.pie(y, labels = labels)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([50, 56, 70, 75])

labels = ["橘子", "苹果", "香蕉", "哈密瓜"]
explode = [0.2, 0, 0, 0]

plt.pie(y, labels = labels,explode = explode,shadow = True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([50, 56, 70, 75])

labels = ["橘子", "苹果", "香蕉", "哈密瓜"]
explode = [0.2, 0, 0, 0]

plt.pie(y, labels = labels,explode = explode,shadow = True)
plt.legend()
plt.show()
```
