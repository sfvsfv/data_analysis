```python
import pandas as pd
data=pd.read_csv("wine.csv")
data
```

```python
x = data.drop('target',axis=1)# 按照列剔除
y = data['target']
x
```

```python
import matplotlib.pyplot as plt
plt.rcParams["font.sans-serif"]=["SimHei"] #设置字体
print(data.target.value_counts())
data.target.value_counts().plot(kind='bar')# 绘制三种酒类的数量柱形图
plt.title('类标签分布')
plt.xlabel("类标签")
plt.ylabel("数量")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler #导入标准化模块
sc = StandardScaler()# 实例化
X_train = sc.fit_transform(X_train) # 训练集标准化
X_test = sc.transform(X_test) # 测试集也需要标准化
X_train # 查看训练集标准化的结果
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 2) # 将13维降为2维
 
X_train = pca.fit_transform(X_train)# 训练集标准化
X_test = pca.transform(X_test) # 测试集标准化
# 降维后的各主成分的方差值占总方差值的比例，即方差贡献率
print("方差贡献率率：", pca.explained_variance_ratio_)
# 降维后各主成分的方差
print("主成分方差：", pca.explained_variance_)
print("主成分个数为：",pca.n_components_) # 返回 PCA 模型保留的主成份个数
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components =0.90) # 阙值95%
X_train = pca.fit_transform(X_train)# 训练集标准化
X_test = pca.transform(X_test) # 测试集标准化
pca.fit(X_train)
print("方差贡献率率：", pca.explained_variance_ratio_)
print("主成分方差：", pca.explained_variance_)
print("主成分个数为：",pca.n_components_) # 返回 PCA 模型保留的主成份个数
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)

fig, ax = plt.subplots(figsize = (18,12))
plot_tree(clf, ax=ax, fontsize = 8, feature_names= x.columns)
plt.savefig('./PCA.jpg', dpi=500, bbox_inches='tight') # 清晰保存
```

```python
from sklearn.metrics import classification_report
y_pred = clf.predict(X_test)#预测
print(classification_report(y_test, y_pred))#评价模型
```
