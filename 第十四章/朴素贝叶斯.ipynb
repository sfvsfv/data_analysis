## 鸢尾花分类

```python
import pandas as pd

data=pd.read_csv('Iris.csv',encoding='gbk')

data
```

```python
x=data.drop('种类',axis=1) #提取自变量,axis=1表示以列删除；默认删除行
y=data['种类'] # 提取目标变量
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0) #分割数据
```

```python
X_test # 查看测试机数据
```

```python
#朴素贝叶斯是一种针对二元（二类）和多类分类问题的分类算法。当使用二进制或分类输入值进行描述时，该技术最容易理解。
from sklearn.naive_bayes import GaussianNB # 导入高斯贝叶斯模型

model = GaussianNB() # 构建模型

model.fit(X_train,y_train) # 拟合训练

predict=model.predict(X_test) # 预测我们的测试集

predict # 查看预测结果
```

```python
from sklearn.metrics import accuracy_score # 导入评估函数

accuracy_score=accuracy_score(y_test,predict) 

print("测试集准确度为%.2f%%："%(100 *accuracy_score))
```

```python
# 打印出模型的分类分数
from sklearn.metrics import classification_report

print(classification_report(y_test, predict))
```

## 多项式利贝叶斯

```python
# DictVectorizer将特征值映射列表转换为向量：当特征值是字符串类型时，此转换器只会执行二进制 one-hot 编码
# 样本中未出现的特征（映射）将在结果数组/矩阵中具有零值
from sklearn.feature_extraction import DictVectorizer 
# 第一个为城市，第二个位农村
data = [
{'房子': 100, '街': 50, '店铺': 25, '汽车': 100, '树木': 20},
{'房子': 5, '街': 5, '店铺': 1, '汽车': 10, '数目': 500, '河流': 1}
] 
dv = DictVectorizer(sparse=False) # 转为向量
dv
```

```python
X = dv.fit_transform(data) # 学习特征名称列表 -> 索引映射并转换 X
X
```

```python
import numpy as np
Y = np.array([1, 0]) #只有0和1两情况.1对应城市，0对应农村
Y
```

```python
from sklearn.naive_bayes import MultinomialNB # 导入多项式分布贝叶斯模型

model = MultinomialNB() # 模型实例化

model.fit(X, Y) # 模型训练
```

```python
test_data = [
{'房子': 90, '街': 50, '店铺': 20, '汽车': 90, '树木': 20,"河流":1},
{'房子': 5, '接': 5, '店铺': 2, '汽车': 10, '树木': 400, '河流': 0}
] 

model.predict(dv.fit_transform(test_data)) # 测试的字符类型也需要转换位向量
# 这个预测是正确的！
```

```python
from sklearn.metrics import accuracy_score # 导入评估函数
accuracy_score=accuracy_score(y_test,predict) 
print("测试集准确度为%.2f%%："%(100 *accuracy_score))
```

## 癌症预测

```python
import pandas as pd

data=pd.read_csv("cancer.csv")

data.head()
```

```python
data.drop(['Unnamed: 32','id'], axis = 1, inplace = True) # inplace用于强制修改覆盖
data
```

```python
diagnosis = data['diagnosis'] # 读取diagnosis属性列
diagnosis
```

```python
# 对目标字符编码为数字
target = pd.get_dummies(diagnosis, drop_first = True) # 使用get_dummies对这一列值编码
target
```

```python
cancer = data.drop('diagnosis', axis = 1)# 先删除目标
cancer
```

```python
new_data=pd.concat([cancer, target], axis = 1) # 合并列
new_data['target'] = new_data['M'] # 把M列复制到新的一列target
new_data.drop('M', axis = 1, inplace = True) # 把M这一列删除
new_data
```

```python
import seaborn as sns
sns.countplot(new_data['target']) # 绘制良性和恶性柱形图
```

```python
sns.scatterplot(x = 'texture_mean', y ='radius_mean', hue = 'target', data = new_data)# 这里人意以纹理平均值与半径平均值绘制
```

```python
new_data.describe()
```

```python
from sklearn.preprocessing import MinMaxScaler #导入缩放器
scaler = MinMaxScaler(feature_range =(0,1)) #实例化缩放器
```

```python
X = new_data.iloc[:,: -1] #读取除了最后一列target的数据
new_x = scaler.fit_transform(X)# 先对x数据缩放
new_x  
```

```python
y = new_data.iloc[:,-1]
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 5)# 切割数据
```

```python
from sklearn.svm import SVC # 导入模型

model = SVC()# 模型实例化
model.fit(X_train,y_train)# 训练模型 
y_pred = model.predict(X_test)# 使用模型预测
y_pred
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
