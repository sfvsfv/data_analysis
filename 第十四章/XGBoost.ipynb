```python
import pandas as pd
data=pd.read_csv('HousingData.csv')
data.head()
```

```python
data.info()
```

```python
(data==0).sum() # 值为0数量
```

```python
data.describe()# 数据描述
```

```python
house = data.dropna() #删除缺失值
house.isnull().sum() #再查看缺失数量
```

```python
house
```

```python
x=house.drop('MEDV',axis=1)
y=house['MEDV']
x
```

```python
#拆分训练集和测试集
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size =0.2,random_state = 0)
```

```python
import xgboost as xgb #导入模型
data_dmatrix = xgb.DMatrix(data=xtrain,label=ytrain) # 参数分别填写xtrain和ytrain
# 模型参数填充
model= xgb.XGBRegressor(objective ='reg:linear', colsample_bytree = 0.3,
        learning_rate = 0.1,max_depth = 5, alpha = 10, n_estimators = 10)
```

```python
# 模型参数填充
model.fit(xtrain,ytrain)#拟合/训练
preds = model.predict(xtest)# 预测
preds# 查看预测结果
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(ytest, preds))#计算均方误差
print("误差均方根: %f" % (rmse))
```

```python
from xgboost import plot_importance
from matplotlib import pyplot as plt
# 显示重要特征,按顺序排名
plot_importance(model)
plt.savefig('./XGBoost排名.jpg', dpi=500, bbox_inches='tight') # 清晰保存
plt.show()
```

```python
node_params = {
    'shape': 'box',
    'style': 'filled,rounded',
    'fillcolor': '#78bceb'
}
#需要到：https://graphviz.org/download/ 下载window版本graphviz并配置环境变量
#同时bin目录的dot.exe添加到环境变量。配置好后重启软件，感兴趣读者可以自行尝试。
xgb.to_graphviz(model, condition_node_params = node_params)
```
